# Axes Labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
```

```python
#setting legend property

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend()
```

```python
#setting legend property

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend(loc='upper right', fontsize="small", shadow=True)  #default loc = 'best'
```

```python
#setting legend property

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend(loc='upper right', fontsize="small", shadow=True)  #default loc = 'best'
plt.grid()  #accurate datapoints
```

# Bar Chart

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.legend()
```

```python
#multiple bars
```

```python
profit = [40, 2, 34, 12]
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, label='Revenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
```

```python
#horizontal bar chart
plt.yticks(xpos, company)
plt.xlabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos+0.2, profit, height=0.4, label='Profit')
plt.legend()
```
