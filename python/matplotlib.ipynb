# Axes Labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
#setting legend property

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend()
```

```python
#setting legend property

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend(loc='upper right', fontsize="small", shadow=True)  #default loc = 'best'
```

# Bar Chart

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.legend()
```

```python
#multiple bars
```

```python
profit = [40, 2, 34, 12]
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, label='Revenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
```

```python
#horizontal bar chart
plt.yticks(xpos, company)
plt.xlabel('Revenue(Bln)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos+0.2, profit, height=0.4, label='Profit')
plt.legend()
```

# Histograms

In histograms, x axis contains a variable and y axis will be a frequency of that variable.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = 3, rwidth=0.8)  #by default no. of bins is set to 10
```

```python
plt.hist(blood_sugar, rwidth=0.5, bins=4)
```

```python
#80-100: Normal
#100-125: Pre-diabetic
#125-150: Diabetic
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color='y')
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')
```

```python
#Multiple Data
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'])
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')
plt.legend()
```

```python
#horizontal orientation
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, orientation='horizontal')
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')
```

# Pie Chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#pie chart to track down home expenses
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5)
plt.show()
```

```python
#explode
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5, explode=[0.1, 0, 0.3, 0, 0.5])
plt.show()
```

```python
#angle properties
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5, startangle=45)
plt.show()
```

```python
#counterclock
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5, counterclock=False, startangle=45)
plt.show()
```

```python
#save chart to a file using savefig
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5, counterclock=False, startangle=45, explode=[0.1, 0, 0.3, 0, 0.5])
plt.savefig("piechart.jpg", bbox_inches='tight', pad_inches=1)
```

```python
#save chart to a file using savefig
plt.pie(exp_vals, labels=exp_labels, shadow=True, autopct='%.2f%%', radius=1.5, counterclock=False, startangle=45, explode=[0.1, 0, 0.3, 0, 0.5])
plt.savefig("piechart.pdf", bbox_inches='tight', pad_inches=1)
```

# Stack Plot

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
```

# Box Plot

```python
#box plot - displaying the distribution of data based on a five number summary
#(minimum, first quartile (Q1), median, third quartile (Q3), and maximum)
one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
two = [1, 2, 3, 4, 5, 4, 3, 2, 1]
three = [6, 7, 8, 9, 6, 7, 8, 3, 4]

data = [one, two, three]
```

```python
plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
two = [1, 2, 3, 4, 5, 4, 3, 2, 1]
three = [6, 7, 8, 9, 6, 7, 8, 3, 4]

data = [one, two, three]
```

```python
plt.violinplot(data, showmedians=True)
plt.show()
```
